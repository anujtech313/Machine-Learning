Principal Component Analysis

It is the method of reducing datasets with higher dimension into lower dimension
that still contains alot of information

```python
# Shifting the values in a mean centered values and standard deviation =1
# As we can see mean value is not equal to 0
data.describe()
```

```python
# Now mean = 0 and std = 1
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scale_data = scalar.fit_transform(data)
```

```python
#If we wish to convert this transformed data to dataframe with original data
df = pd.DataFrame(scale_data, columns = data.columns)
```

```python
# e = -14 is near to 0. We have fulfilled the first rule of pca for converting the data into mean = 0 and std = 1
df.describe()
```
